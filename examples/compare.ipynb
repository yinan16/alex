```python
from alex.alex import dsl_parser
from alex.alex import compare
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 600
```

```python
config_path_1 = "../examples/configs/small1.yml"
config_path_2 = "../examples/configs/small2.yml"    
```

---
**Note**

- This is not merely the difference between two yaml files. The yaml file is a Domain Specific Language (DSL) used to describe the network. 
- The network diff is the diff between two abstract syntax trees parsed from the DSL.  

---


```python
render_diff_path = ""
compare.diff(config_path_1, config_path_2, render_to=render_diff_path)
```

```python
render_dist_path = ""
compare.dist(config_path_1, config_path_2, render_to=render_dist_path)
```
